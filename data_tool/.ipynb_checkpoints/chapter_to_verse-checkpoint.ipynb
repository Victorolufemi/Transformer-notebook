```python
import re

#This script converts all the chapters of the english and yoruba bible text to verses.
```

```python
from pathlib import Path

DATA_DIR = "./data/"
d = Path(DATA_DIR)
english_verses = []
igbo_verses = []
```

```python
def convert_file_into_array(file_name):
    file_list = []
    with open(file_name, "r", encoding="utf-8") as file:
        for line in file:
            file_list.append(line)
    return file_list
```

```python
english_chapters = convert_file_into_array(DATA_DIR+"english-chapters.txt")
yoruba_chapters = convert_file_into_array(DATA_DIR+"yoruba-chapters.txt")
```

```python
def get_verses(first_chapters, second_chapters):
    first_verses = []
    second_verses = []
    for first, second in zip(first_chapters, second_chapters):
        first_list = re.split("(\(\d+\))", first)
        second_list = re.split("(\(\d+\))", second)
        if len(first_list) != len(second_list):
            continue
        first_verses.extend(get_verse(first_list))
        second_verses.extend(get_verse(second_list))
    return first_verses, second_verses
```

```python
def get_verse(chapter):
    begin = ""
    verse = []
    for word in chapter:
        if word.strip() == "":
            continue
        if len(word) <= 6:
            word = word.replace("(", "")
            begin = word.replace(")", "")
        else:
            begin += word.replace("\n", "")
            verse.append(begin.strip())
    return verse
```

```python
english_verses, yoruba_verses = get_verses(english_chapters, yoruba_chapters)
```

```python
def save_file(array, file_name):
    with open(DATA_DIR+file_name, "w+", encoding="utf-8") as file:
        joined_array = "\n".join(array)
        file.write(joined_array)
```
